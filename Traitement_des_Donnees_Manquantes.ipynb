### Création d'une fonction généralisée pour tout dataframe contenant des données manquantes

```python
def percent_missing(df):
  percent_nan = 100 * df.isnull().sum() / len(df)
  percent_nan = percent_nan[percent_nan > 0].sort_values(ascending=False)
  
  return percent_nan
```

```python
percent_nan = percent_missing(df)
```

# Suppression des Features ou Suppression des lignes

Si quelques lignes seulement, par rapport à la taille de notre ensemble de données, manquent de valeurs, il est peut être judicieux d'éliminer ces lignes. Qu'est-ce que cela nous coûte en termes de performance ? Cela supprime essentiellement des données d'entraînement/de test potentielles, mais s'il ne s'agit que de quelques lignes, il est peu probable que cela change les performances.

Parfois, c'est une bonne idée de supprimer entièrement une Feature (caractéristique) si elle a trop de valeurs manquantes (null). Cependant, vous devez examiner attentivement pourquoi il y a autant de valeurs manquantes. Dans certaines situations, les valeurs manquantes pourraient simplement être utilisées comme une catégorie distincte. 

Prenons l'exemple d'une colonne de features indiquant le nombre de voitures que peut accueillir un garage. S'il n'y a pas de garage, il y a peut-être une valeur manquante (null) au lieu d'un zéro. Dans ce cas, il est surement plus logique de remplir rapidement les valeurs manquantes avec un 0 plutôt qu'une valeur null. Vous seul pouvez décider en fonction de votre expertise du domaine et de votre connaissance de l'ensemble de données !

## Remplir ou supprimer des données ?

Explorons comment choisir de supprimer ou de compléter les données manquantes pour les lignes auxquelles il manque des données. Choisissons un seuil à partir duquel nous décidons qu'il est acceptable de supprimer une ligne s'il lui manque des données (au lieu d'essayer de compléter ce point de données manquant). Nous choisirons 1% comme seuil. Cela signifie que si moins de 1% des lignes manquent pour cette feature, nous envisagerons d'abandonner cette ligne, au lieu de traiter la feature elle-même. Il n'y a pas de bonne réponse ici, il suffit d'utiliser notre bon sens et nos connaissances du domaine de l'ensemble de données, évidemment nous ne voulons pas nous débarasser d'un seuil très élevé comme 50%, nous devrions également explorer la corrélation à l'ensemble de données, peut-être est-il logique de supprimer la feature à la place.

En nous basant sur la description textuelle des features, nous verrons que la plupart de ces données manquantes sont en fait des NaN intentionnellement utilisés pour remplacer 0 ou "None".

## Suppression des lignes

Mais ici nous allons montrer comment les supprimer au cas où vous vous trouvez dans une situation où il est plus logique de se débarasser d'une ligne, sur la base de features de colonnes manquantes.

# [Imputation](https://en.wikipedia.org/wiki/Imputation_(statistics)) sur les Données Manquantes

Pour imputer les données manquantes, nous devons décider quelle autre feature bien complétée (pas de valeurs NaN) est la plus corrélée aux données de la feature manquante. Dans ce cas particulier, nous utiliserons :

"Neighborhood" : Emplacements physiques (quartiers) dans les limites de la ville d'Ames

"Lot Frontage" : Longueur (en pied) de la rue devant la propriété

Nous partons du principe que la longeur du lot (Lot Frontage) est liée au quartier (Neighborhood) dans lequel se trouve la maison.

```python
df['Neighborhood'].unique()
```

```python
plt.figure(figsize=(8, 12))
sns.boxplot(x='Lot Frontage', y='Neighborhood', data=df, orient='h');
```

## Imputatons les données manquantes en fonction d'autres features

Il existe des méthodes plus complexes, mais en général, plus c'est simple, mieux c'est. Cela évite de construire des modèles par-dessus d'autres modèles.

Plus d'informations sur les options : https://scikit-learn.org/stable/modules/impute.html

```python
df.groupby('Neighborhood')['Lot Frontage'].mean()
```

## Transformation de colonne

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.transform.html

```python
df.head()['Lot Frontage']
```

```python
df[df['Lot Frontage'].isnull()]
```

```python
df.iloc[21:26]['Lot Frontage']
```

```python
df.head()
```

```python
df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean()))
```

```python
df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean())).iloc[21:26]
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean()))
```

```python
percent_nan = percent_missing(df)
percent_nan
```

```python
sns.barplot(x=percent_nan.index, y=percent_nan)
plt.xticks(rotation=90);
```

```python
df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
```

```python
percent_nan = percent_missing(df)
percent_nan
```
